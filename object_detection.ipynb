```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load Image
img = cv2.imread("football1.jpg")  # Ensure image is uploaded
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print("Football", img.shape)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Split Channels
B, G, R = cv2.split(img)
print('BLUE', B.shape)
plt.imshow(B, cmap='gray')
plt.axis("off")
plt.show()

print('GREEN', G.shape)
plt.imshow(G, cmap='gray')
plt.axis("off")
plt.show()

print('RED', R.shape)
plt.imshow(R, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# Convert to Grayscale
grayImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(grayImg, cmap='gray')
plt.axis("off")
plt.show()
print(grayImg.shape)
```

```python
# Convert to HSV
hsvImage = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(hsvImage)
plt.axis("off")
plt.show()
print(hsvImage.shape)
```

```python
# Convert to YCrCb
YCRImage = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
plt.imshow(YCRImage)
plt.axis("off")
plt.show()
print(YCRImage.shape)
```

```python
# Convert to LAB
LabImage = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
plt.imshow(LabImage)
plt.axis("off")
plt.show()
print(LabImage.shape)
```

```python
# Edge Detection
sobelXedge = cv2.Sobel(grayImg, cv2.CV_64F, 1, 0, ksize=5)
sobelYedge = cv2.Sobel(grayImg, cv2.CV_64F, 0, 1, ksize=5)
SobelXYedge = cv2.Sobel(grayImg, cv2.CV_64F, 1, 1, ksize=5)

print('Sobel X-Order Edge')
plt.imshow(sobelXedge, cmap='gray')
plt.axis("off")
plt.show()

print("Sobel Y-Order Edge")
plt.imshow(sobelYedge, cmap='gray')
plt.axis("off")
plt.show()

print('Sobel XY-Order Edge')
plt.imshow(SobelXYedge, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# Canny Edge Detection
edges = cv2.Canny(grayImg, threshold1=30, threshold2=100)
edges1 = cv2.Canny(grayImg, threshold1=50, threshold2=150)
edges2 = cv2.Canny(hsvImage[:, :, 2], threshold1=70, threshold2=200)

print('Canny Edge-30-100')
plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.show()

print('Canny Edge-50-150')
plt.imshow(edges1, cmap='gray')
plt.axis("off")
plt.show()

print('Canny Edge-70-200')
plt.imshow(edges2, cmap='gray')
plt.axis("off")
plt.show()
```
